<h1>K-Means</h1>

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

student_data = pd.read_csv("../Datasets/turkiye-student-evaluation_generic.csv")
df = student_data.iloc[:,5:33]

# Transformation of data through PCA
pca = PCA(n_components = 2, random_state=1)
df_pca = pca.fit_transform(df)

# set number of clusters
kclusters = 3

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(df_pca)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:50]

y_kmeans = kmeans.predict(df_pca)

# Ploting clusters
plt.figure(figsize=(10, 7))
plt.scatter(df_pca[:, 0], df_pca[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
```
