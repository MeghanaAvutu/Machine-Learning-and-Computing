<h1>Fizzy C-Means</h1>

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from fcmeans import FCM
import matplotlib.pyplot as plt

student_data = pd.read_csv("../Datasets/turkiye-student-evaluation_generic.csv")
df = student_data.iloc[:,5:33]

# Transformation of data through PCA
pca = PCA(n_components = 2, random_state=1)
df_pca = pca.fit_transform(df)

# fit the fuzzy-c-means
fcm = FCM(n_clusters=3)
fcm.fit(df_pca)

# outputs
fcm_centers = fcm.centers
fcm_labels  = fcm.u.argmax(axis=1)

y_cmeans = fcm.predict(df_pca)
y_cmeans

plt.figure(figsize=(10, 7))
plt.scatter(df_pca[:, 0], df_pca[:, 1], c=y_cmeans, s=50, cmap='viridis')
```
